```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


class InsurancePrediction:

    def __init__(self, file_path: str) -> None:
        self.file_path = file_path
        self.df = None
        self.model = LogisticRegression()
        self.x_train = None
        self.x_test = None
        self.y_train = None
        self.y_test = None

    def load_data(self) -> None:
        self.df = pd.read_csv(self.file_path)
        print("Data loaded.\n")

    def verify_data(self) -> None:
        print("Shape:", self.df.shape)
        print("\nHead:\n", self.df.head())
        print("\nInfo:")
        self.df.info()
        print("\nMissing values:\n", self.df.isnull().sum())

    def process_data(self) -> None:
        if 'Unnamed: 0' in self.df.columns:
            self.df.drop('Unnamed: 0', axis=1, inplace=True)

    def split_data(self) -> None:
        x = self.df[['age']]
        y = self.df['bought_insurance']

        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(x, y, test_size=0.2, random_state=2)

    def visualize_data(self) -> None:
        plt.scatter(self.df['age'], self.df['bought_insurance'], marker='o')
        plt.xlabel("Age")
        plt.ylabel("Bought Insurance (0/1)")
        plt.title("Age vs Insurance Purchase")
        plt.show()

    def train_model(self) -> None:
        self.model.fit(self.x_train, self.y_train)
        print("Model trained.\n")

    def test_model(self) -> np.ndarray:
        return self.model.predict(self.x_test)

    def evaluate_model(self) -> None:
        y_pred = self.test_model()

        print("Coefficient:", self.model.coef_[0][0])
        print("Intercept:", self.model.intercept_[0])
        print("Accuracy:", accuracy_score(self.y_test, y_pred))
        print("\nConfusion Matrix:\n", confusion_matrix(self.y_test, y_pred))
        print("\nClassification Report:\n", classification_report(self.y_test, y_pred))

    def plot_sigmoid_curve(self) -> None:
        x_values = np.linspace(self.df['age'].min(), self.df['age'].max(), 100)
        z = self.model.coef_[0][0] * x_values + self.model.intercept_[0]
        y_prob = 1 / (1 + np.exp(-z))

        plt.scatter(self.df['age'], self.df['bought_insurance'])
        plt.plot(x_values, y_prob)
        plt.xlabel("Age")
        plt.ylabel("Probability of Buying Insurance")
        plt.title("Sigmoid Curve")
        plt.show()

    def predict_insurance(self, age: float) -> int:
        return self.model.predict([[age]])[0]

if __name__ == "__main__":
    pipeline = InsurancePrediction("insurance_data.csv")

    pipeline.load_data()
    pipeline.verify_data()
    pipeline.process_data()
    pipeline.visualize_data()
    pipeline.split_data()
    pipeline.train_model()
    pipeline.evaluate_model()
    pipeline.plot_sigmoid_curve()

    age = float(input("Enter age: "))
    result = pipeline.predict_insurance(age)
    print("Will Buy Insurance" if result == 1 else "Will NOT Buy Insurance")
```
