```python
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression

X = df[['Hours Studied', 'Sample Question Papers Practiced']].astype(float)
y = df['Performance Index'].astype(float)

reg = LinearRegression()
reg.fit(X, y)

x1_range = np.linspace(X['Hours Studied'].min(), X['Hours Studied'].max(), 50)
x2_range = np.linspace(
    X['Sample Question Papers Practiced'].min(),
    X['Sample Question Papers Practiced'].max(),
    50
)

X1, X2 = np.meshgrid(x1_range, x2_range)

Z = (
    reg.coef_[0] * X1 +
    reg.coef_[1] * X2 +
    reg.intercept_
)

# ---------------------------
# 4. Plot 3D scatter + plane
# ---------------------------
fig = go.Figure()

# Data points
fig.add_trace(go.Scatter3d(
    x=X['Hours Studied'],
    y=X['Sample Question Papers Practiced'],
    z=y,
    mode='markers',
    marker=dict(size=5),
    name='Actual Data'
))

# Best-fit plane
fig.add_trace(go.Surface(
    x=X1,
    y=X2,
    z=Z,
    colorscale='Viridis',
    opacity=0.8,
    name='Best Fit Plane'
))

# ---------------------------
# 5. Layout settings
# ---------------------------
fig.update_layout(
    title='Multiple Linear Regression: Best Fit Plane',
    width=1000,
    height=800,
    scene=dict(
        xaxis_title='Hours Studied',
        yaxis_title='Sample Question Papers Practiced',
        zaxis_title='Performance Index',
        xaxis=dict(range=[X['Hours Studied'].min(), X['Hours Studied'].max()]),
        yaxis=dict(range=[
            X['Sample Question Papers Practiced'].min(),
            X['Sample Question Papers Practiced'].max()
        ]),
        zaxis=dict(range=[y.min(), y.max()])
    )
)

fig.show()
```

```python
#Linear_Regression Code using Oops concept

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


class SalaryPredictionModel:
    def __init__(self, file_path: str) -> None:
        self.file_path = file_path
        self.df = None
        self.model = LinearRegression()
        self.x_train = None
        self.x_test = None
        self.y_train = None
        self.y_test = None

    def load_data(self) -> None:
        self.df = pd.read_csv(self.file_path)
        print("Data loaded")

    def verify_data(self) -> None:
        print("First 5 rows:\n", self.df.head())
        print("\nData Info:\n")
        print(self.df.info())
        print("\nMissing values:\n", self.df.isnull().sum())

    def process_data(self) -> None:
        if 'Unnamed: 0' in self.df.columns:
            self.df.drop('Unnamed: 0', axis=1, inplace=True)

    def split_data(self) -> None:
        x = self.df[['YearsExperience']]
        y = self.df['Salary']
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(x, y, test_size=0.2, random_state=2)

    def visualize_data(self) -> None:
        plt.scatter(self.df['YearsExperience'], self.df['Salary'], marker='o')
        plt.xlabel('Years of Experience')
        plt.ylabel('Salary')
        plt.title('Experience vs Salary')
        plt.show()

    def train_model(self) -> None:
        self.model.fit(self.x_train, self.y_train)
        print("Model training completed.\n")

    def validate_model(self) -> None:
        plt.scatter(self.x_train, self.y_train)
        plt.plot(self.x_train, self.model.predict(self.x_train), color='red')
        plt.title('Training Data with Regression Line')
        plt.show()

    def test_model(self) -> np.ndarray:
        return self.model.predict(self.x_test)

    def evaluate_model(self) -> None:
        y_pred = self.test_model()
        print("Model Coefficient (m):", self.model.coef_[0])
        print("Model Intercept (b):", self.model.intercept_)
        print("R2 Score:", r2_score(self.y_test, y_pred))
        print("MSE:", mean_squared_error(self.y_test, y_pred))

    def predict_salary(self, years: float) -> float:
        prediction = self.model.predict([[years]])
        return prediction[0]

if __name__ == "__main__":
    pipeline = SalaryPredictionModel("Salary_dataset.csv")

    pipeline.load_data()
    pipeline.verify_data()
    pipeline.process_data()
    pipeline.visualize_data()
    pipeline.split_data()
    pipeline.train_model()
    pipeline.validate_model()
    pipeline.evaluate_model()

    years_exp = float(input("Enter Years of Experience: "))
    salary = pipeline.predict_salary(years_exp)
    print(f"Predicted Salary: {salary:.2f}")

```
