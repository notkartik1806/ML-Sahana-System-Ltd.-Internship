```python
my_int_list = []
my_str_list = []


while (1):
    inp = (input("enter anything: "))
    if inp.isnumeric():
        if int(inp) == 0:
            break
        my_int_list.append(int(inp))
    elif inp.isalpha():
        my_str_list.append(inp)

print(my_int_list)
print(my_str_list)


print("minimum in int list is: {}".format(min(my_int_list)))
print("maximum in int list is: {}".format(max(my_int_list)))

my_str_list.reverse()

print(my_str_list)
```

```python
my_dict = {}

while (1):
    inp = input("enter student number: ")
    if int(inp) == 0:
        break
    my_dict[inp] = {
        "name" : input("enter name: "),
        "age" : int(input("enter age: ")),
        "marks" : int(input("enter marks: "))
    }

    if 90 < my_dict[inp]["marks"] < 100:
        my_dict[inp]["grade"] = "A"
    if 80 < my_dict[inp]["marks"] < 90:
        my_dict[inp]["grade"] = "B"
    if 70 < my_dict[inp]["marks"] < 80:
        my_dict[inp]["grade"] = "C"
    else:
        my_dict[inp]["grade"] = "Fail"


print(my_dict)
```
